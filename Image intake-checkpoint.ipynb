```python
import cv2

def capture_and_save():
    camera = cv2.VideoCapture(0)  # Open default camera (index 0)
    
    while True:
        _, image = camera.read()  # Capture frame-by-frame
        cv2.imshow('Text Detection', image)  # Display the captured frame
        
        # Press 's' key to save the image and exit loop
        if cv2.waitKey(1) & 0xFF == ord('s'):
            cv2.imwrite('Test1.jpg', image)  # Save the captured image
            break
    
    camera.release()  # Release the camera
    cv2.destroyAllWindows()  # Close all OpenCV windows

capture_and_save()
```
